```python
# https://www.kaggle.com/datafiniti/hotel-reviews
import pandas as pd

df = pd.read_csv('Datafiniti_Hotel_Reviews.csv')

categories = ['name','city','country','reviews.rating','reviews.title','reviews.text','reviews.userCity']
data = df[categories].fillna('0')

from sklearn.model_selection import train_test_split

train,test = train_test_split(data.values, test_size=0.7)

```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer
from sklearn import metrics
import numpy as np

vectorizer = CountVectorizer()

train_features = vectorizer.fit_transform([x[5] for x in train])
test_features = vectorizer.transform([x[5] for x in test])

# This will train the model using the word counts we computed and the existing classifications in the training set
model = MultinomialNB()
model.fit(train_features, [int(r[3]) for r in train])

# Now we can use the model to predict classifications for our test features
predictions = model.predict(test_features)

predictions


test2 = pd.DataFrame(test)
test2['predictions'] = np.array(predictions)
test2
```

```python
from sklearn.metrics import accuracy_score
accuracy_score([int(r[3]) for r in test], predictions)
```
